```python
import pandas
cmr=pandas.read_csv("http://modcom.co.ke/bigdata/datasets/Customers.csv")
cmr
```

```python
print(cmr.corr())
print(cmr.describe())

array = cmr.values
x=array[:, 2:6]
y=array[:, 6]

from sklearn import model_selection
x_train,x_test, y_train, y_test  = model_selection.train_test_split(x,y,
                                                                     test_size=0.30,
                                                                     random_state=42)

from sklearn.linear_model import LinearRegression
model =LinearRegression ()
model.fit(x_train, y_train) 

predictions=model.predict(x_test)
print (predictions)
from sklearn.metrics import  r2_score
print (r2_score(y_test,  predictions))
amount=model.predict([[32,11,35,2]])
print(amount)
```

```python
import pandas
cmr1=pandas.read_csv("http://modcom.co.ke/bigdata/datasets/Customers.csv")
cmr1

```

```python
#machine learning does not support text in the features e.g avatar
#you have to encode any categorical features to 0, 1,2.......
#you can encode in two ways :Manually and automatically
#automatically is the better choice
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

cmr1['AvatarEncoded'] =encoder.fit_transform(cmr1['Avatar'])
#subset = cmr1[['AvatarEncoded','Avatar']]
#print(subset)
print(cmr1['Avatar'])

array = cmr1.values
x=array[:, 2:6]
y=array[:, 6]

from sklearn import model_selection
x_train,x_test, y_train, y_test  = model_selection.train_test_split(x,y,
                                                                     test_size=0.30,
                                                                     random_state=42)

from sklearn.linear_model import LinearRegression
model =LinearRegression ()
model.fit(x_train, y_train) 

predictions=model.predict(x_test)
print (predictions)
from sklearn.metrics import  r2_score
print (r2_score(y_test,  predictions))


```
